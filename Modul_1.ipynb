УСЛОВИЯ ЗАДАНИЯ:

Компьютер загадывает целое число от 1 до 100, и нам его нужно угадать. Под «угадать» подразумевается «написать программу, которая угадывает число».
Алгоритм учитывает информацию о том, больше или меньше случайное число нужного нам числа.

```python
# импорт библиотеки numpy для работы со случайными числами
import numpy as np 

def guess_the_number(number:int=1) -> int:
    ''' Функция, получающая на вход число, а на выходе 
    выдающая количество попыток его угадать.
    Алгоритм функции заключается в сокращении диапазона,
    в котором может находится функция, путём сравнения
    загаданного числа с числом, находящимся в середине 
    диапазона '''
    
    count = 0 # счётчик
    min_v = 1 # минимальное значение диапазона
    max_v = 101 # максимальное значение диапазона, увеличенное на 1

    while True:
        count+=1 #увеличение значения счётчика

        ''' устанавливаем вариант ответа, как
        среднее арифмитическое между началом
        и концом диапазона, с отбрасыванием
        дробной части. Для чисел, между которыми
        только 1 число делается исключение
        '''
        if (max_v - min_v > 1):
            predict = (max_v + min_v)//2
        else:
            predict = max_v - 1
                
        if predict == number:
            return(count)
            break # если число угадано, возвращаем результат и покидаем цикл
        elif number > predict:
            min_v = predict # если загадонное число больше, нижней границей устанавливаем вариант ответа
        else:
            max_v = predict # если загаданное число меньше, yстанавливаем верхней границей вариант ответа

           
list_attempt = [] # создаём пустой список, в который будем собирать варианты возвращаемых данных
 
# цикл который 1000 раз загадывает и угадывает число 
for i in range(1000):
    rand_number = np.random.randint(1,101) # присваиваем случайное значение загаданному числу
    list_attempt.append(guess_the_number(rand_number)) # запускаем функцию и добавляем полученное количество попыток в лист

# вывод на экран максимального количества попыток из полученного списка
print(f'Максимальное количество попыток угадать на 1000 игр: {max(list_attempt)}')  
```

Вывод:<p>
Представленное решение обеспечивает угадывание числа от 1 до 100 менее, чем за 7 попыток.
